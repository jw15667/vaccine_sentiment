```python
import re

def cleaned_headline(sentence):
    new_sent = re.sub("\n", "", sentence)
    return new_sent

def cleaned_headline_2(sentence):
    new_sent = re.sub("\xa0", " ", sentence)
    return new_sent

def make_sentence(sentence):
    sent = sentence.split(' ')
    word = [word.lower() for word in sent]
    return word
```

```python
telegraph['new_title'] = telegraph['title'].apply(cleaned_headline)
```

```python
telegraph['new_title'] = telegraph['new_title'].apply(cleaned_headline_2)
```

```python
telegraph['no_punc_title'] = telegraph.new_title.apply(lambda x: re.sub(r'[^\w\s]', '',x))
```

```python
telegraph['new_title'] = telegraph['new_title'].apply(make_sentence)
```

```python
news_dict = {}
for headline in words_list:
    f = headline.split(' ')
    for word in f:
        x = word.lower()
        if x not in stop_words:
            if x not in news_dict:
                news_dict[x] = 1
            else:
                news_dict[x] += 1
```

```python
 sort_telegraph = sorted(news_dict.items(), key=lambda x: x[1], reverse=True)
```

```python
guardian['new_web_title'] = guardian.web_title.apply(make_sentence)
```

```python
def in_covid_dictionary():
    sent = sentence.split(' ')
    word = [word.lower() for word in sent]
    for x in word:
        if x in sort_telegraph.keys():
            return 'covid_related'
            break
        else:
            return 'n/a'
```

```python
guardian['covid_related'] = guardian['new_web_title'].apply(in_covid_dictionary())
```
