'$(' = "other sentence-internal punctuation mark"

'$,' = "comma"

'$.' = "sentence-final punctuation mark"

'ADJA' = "adjective, attributive"

'ADJD' = "adjective, adverbial or predicative"

'ADV' = "adverb"

'APPR' = "preposition; circumposition left"

'ART' = "definite or indefinite article"

'CARD' = "cardinal number"

'KON' = "coordinate conjunction"

'NE' = "proper noun"

'NN' = "noun, singular or mass"

'PIS' = "substituting indefinite pronoun"

'PRF' = "reflexive personal pronoun"

'PTKNEG' = "negative particle"

'VAFIN' = "finite verb, auxiliary"

'VAINF' = "infinitive, auxiliary"

'VVFIN' = "finite verb, modal"

```python
def is_proper_noun(tag):
    return tag in ['NE']
def is_noun(tag):
    return tag in ['NN']
def is_verb(tag):
    return tag in ['VAFIN', 'VAINF', 'VVFIN']
def is_adverb(tag):
    return tag in ['ADV']
def is_adjective(tag):
    return tag in ["ADJA", 'ADJD']

def penn_to_pos(headline_list):
    tagged_list = []
    
    for tag in headline_list:
        tag = list(tag)
        
        if is_adjective(tag[1]):
            tag[1] = "ADJ"
            tagged_list.append(tag)
        elif is_noun(tag[1]):
            tag[1] = "NOUN"
            tagged_list.append(tag)
        elif is_adverb(tag[1]):
            tag[1] = "ADV"
            tagged_list.append(tag)
        elif is_verb(tag[1]):
            tag[1] = "VERB"
            tagged_list.append(tag)
        elif is_proper_noun(tag[1]):
            tag[1] = 'NE'
            tagged_list.append(tag)
        else:
            continue

    return tagged_list
```

```python
news_dict = {}

for i in list(df_model.index):
    #cleaning words_list for every index
    words_list=[]
    words_list+=df_model["headline"][i].split(' ')
    words_set=set(words_list)
    for x in words_list:
        if x not in news_dict:
            news_dict[x] = 1
        else:
            news_dict[x] += 1
```

```python
news_dict_sort = sorted(news_dict.items(), key=lambda x: x[1], reverse=True)
news_dict_sort
```

```python
keywords=pd.DataFrame()

keywords["keyword"]=[news_dict_sort[i][0] for i in range(len(news_dict_sort))]
keywords["count"]=[news_dict_sort[i][1] for i in range(len(news_dict_sort))]

keywords=keywords[keywords["count"]>25]
```

```python
corona_list=["corona","rki","drosten","wieler","merkel","spahn","coronavirus","virus","lockdown","pandemie","epidemie","pcr-test","impfzentren","hausärzte","schnelltest","teststrategie","intensivbetten","covid","covid-19","impfung","impfstrategie","impfen","biontech","pfizer", "biontech/pfizer","moderna","johnson&johnson", "curevac","astrazeneca","oxford","janssen"]



def corona_related(x):
    temp_list=x.split(' ')

    score=0
    
    for word in temp_list:
        #print(word)
        if word in corona_list:
            score_corona=1
            #print(score)
            return score_corona
        
        elif word in list(keywords["keyword"]):
            score+=1
            #print(score)
        else:
            continue
            
    return score/(len(temp_list))
```

```python
df_model["corona_score"]=df_model["headline"].apply(lambda x: corona_related(x))
```
