```python
#automating request

#uuid=[]
title=[]
href=[]
release_date=[]
#uri=[]
snippet=[]
#supertitle=[]
teaser_text=[]
teaser_title=[]

#keys=list(request["matches"][0].keys())

#automating request

api = zeit.API()
api.set_token(KEY) #put your key in here

for keyword in keywords_zeit:

    print("_________________")
    print(keyword)
    print(api.get_status())
    print("_________________")


    for t in range(len(dates)):

        random_i = randint(5, 10)
        time.sleep(random_i)
        print("loop: ", t)
        print()

        request=api.get(url = "http://api.zeit.de/content", limit = 1000, search = keyword, time_range=dates[t])

        #title += [request["matches"][i]["title"] for i in range(len(request["matches"]))]
        for i in range(len(request["matches"])):
            try:
                title.append(request["matches"][i]["title"])
            except:
                title.append("no title")
                continue
        print("title: ",len(title))

        #release_date += [request["matches"][i]["release_date"] for i in range(len(request["matches"]))]
        for j in range(len(request["matches"])):
            try:
                release_date.append(request["matches"][j]["release_date"])
            except:
                release_date.append("no date")
                continue
        print("date: ",len(release_date))

        #snippet += [request["matches"][i]["snippet"] for i in range(len(request["matches"]))]
        for k in range(len(request["matches"])):
            try:
                snippet.append(request["matches"][k]["snippet"])
            except:
                snippet.append("no snippet")
                continue
        print("snippet: ",len(snippet))

        #teaser_text += [request["matches"][i]["teaser_text"] for i in range(len(request["matches"]))]
        for l in range(len(request["matches"])):
            try:
                teaser_text.append(request["matches"][l]["teaser_text"])
            except:
                teaser_text.append("no teaser text")
                continue
        print("teaser_text: ",len(teaser_text))


        #teaser_title += [request["matches"][i]["teaser_title"] for i in range(len(request["matches"]))]
        for m in range(len(request["matches"])):
            try:
                teaser_title.append(request["matches"][m]["teaser_title"])
            except:
                teaser_title.append("no teaser title")
                continue
        print("teaser_title: ", len(teaser_title))

        for n in range(len(request["matches"])):
            try:
                href.append(request["matches"][n]["href"])
            except:
                href.append("no href")
                continue
        print("href: ", len(href))

        print("-----")
        print()
```

### Scraping data

```python
url = []
headline = []
is_paid = []
date_published = []
channel = []

for date in dates:
    random_i = randint(5, 20)
    time.sleep(random_i)
    
    print()
    print(date)
    
    archive_entries_from_html = spon.archive.scrape_html(spon.archive.html_by_date(date))
    print()
    
    url += [archive_entries_from_html[i]["url"] for i in range(len(archive_entries_from_html))]
    headline += [archive_entries_from_html[i]["headline"] for i in range(len(archive_entries_from_html))]
    is_paid += [archive_entries_from_html[i]["is_paid"] for i in range(len(archive_entries_from_html))]
    date_published += [archive_entries_from_html[i]["date_published"] for i in range(len(archive_entries_from_html))]
    channel += [archive_entries_from_html[i]["channel"] for i in range(len(archive_entries_from_html))]
    print(len(channel))
    
```

```python
spiegel = pd.DataFrame()

spiegel['url'] = [u for u in url]
spiegel['headline'] = [h for h in headline]
spiegel['is_paid'] = [p for p in is_paid]
spiegel['date_published'] = [date for date in date_published]
spiegel['channel'] = [c for c in channel]
```
