```python
keywords_zeit=["corona", "astrazeneca", "biontech", "curevac", "moderna", "johnson&johnson"]
```

```python
#automating request

#uuid=[]
title=[]
href=[]
release_date=[]
#uri=[]
snippet=[]
#supertitle=[]
teaser_text=[]
teaser_title=[]

#keys=list(request["matches"][0].keys())

#automating request

api = zeit.API()
api.set_token(KEY) #put your key in here

for keyword in keywords_zeit:

    print("_________________")
    print(keyword)
    print(api.get_status())
    print("_________________")


    for t in range(len(dates_zeit)):

        random_i = randint(5, 10)
        time.sleep(random_i)
        print("loop: ", t)
        print(dates_zeit[t])
        print()

        request=api.get(url = "http://api.zeit.de/content", limit = 1000, search = keyword, time_range=dates_zeit[t])

        #title += [request["matches"][i]["title"] for i in range(len(request["matches"]))]
        for i in range(len(request["matches"])):
            try:
                title.append(request["matches"][i]["title"])
            except:
                title.append("no title")
                continue
        print("title: ",len(title))

        #release_date += [request["matches"][i]["release_date"] for i in range(len(request["matches"]))]
        for j in range(len(request["matches"])):
            try:
                release_date.append(request["matches"][j]["release_date"])
            except:
                release_date.append("no date")
                continue
        print("date: ",len(release_date))

        #snippet += [request["matches"][i]["snippet"] for i in range(len(request["matches"]))]
        for k in range(len(request["matches"])):
            try:
                snippet.append(request["matches"][k]["snippet"])
            except:
                snippet.append("no snippet")
                continue
        print("snippet: ",len(snippet))

        #teaser_text += [request["matches"][i]["teaser_text"] for i in range(len(request["matches"]))]
        for l in range(len(request["matches"])):
            try:
                teaser_text.append(request["matches"][l]["teaser_text"])
            except:
                teaser_text.append("no teaser text")
                continue
        print("teaser_text: ",len(teaser_text))


        #teaser_title += [request["matches"][i]["teaser_title"] for i in range(len(request["matches"]))]
        for m in range(len(request["matches"])):
            try:
                teaser_title.append(request["matches"][m]["teaser_title"])
            except:
                teaser_title.append("no teaser title")
                continue
        print("teaser_title: ", len(teaser_title))

        for n in range(len(request["matches"])):
            try:
                href.append(request["matches"][n]["href"])
            except:
                href.append("no href")
                continue
        print("href: ", len(href))

        print("-----")
        print()
```

### DATA PREPROCESSING

```python
def data_clean(x):
    x=str(x)
    return x[:10]
```

```python
zeit_df["date_published"]=zeit_df["date_published"].apply(lambda x: data_clean(x))

zeit_df["snippet"]=zeit_df["snippet"].apply(lambda x: re.sub("<em>", "", x))
zeit_df["snippet"]=zeit_df["snippet"].apply(lambda x: re.sub("</em>", "", x))
zeit_df["snippet"]=zeit_df["snippet"].apply(lambda x: re.sub("\n", " ", x))
```

```python
zeit_df["original_headline"]=zeit_df["title"]+" "+zeit_df["snippet"]+" "+zeit_df["teaser_text"]
```

```python
import re
import string

def cleaning_punctuation(x):
    x=str(x)
    x=x.lower()
    remove = string.punctuation
    remove = remove.replace("-", "")
    remove = remove.replace("/", "")
    remove = remove.replace("&", "")# don't remove hyphens
    pattern = r"[{}]".format(remove)
    x=re.sub(pattern, "", x)
    return x

zeit_df["title"]=zeit_df["title"].apply(lambda x: cleaning_punctuation(x))
zeit_df["snippet"]=zeit_df["snippet"].apply(lambda x: cleaning_punctuation(x))
zeit_df["teaser_text"]=zeit_df["teaser_text"].apply(lambda x: cleaning_punctuation(x))
zeit_df["teaser_title"]=zeit_df["teaser_title"].apply(lambda x: cleaning_punctuation(x))
```

```python
import nltk
#nltk.download('stopwords')
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import CountVectorizer

german_stop_words = stopwords.words('german')
vect = CountVectorizer(stop_words = german_stop_words)

def stop_word_removal(x):
    token = x.split()
    return ' '.join([w for w in token if not w in german_stop_words])

cols=['title','snippet','teaser_text','teaser_title']

for col in cols:
    zeit_df[col]=zeit_df[col].apply(lambda x: stop_word_removal(x))
```

```python
import spacy
nlp = spacy.load('de_core_news_sm')

def lemmatizer(x):
    token = nlp(x)
    #x.lemma_
    return ' '.join([t.lemma_ for t in token])

cols=['title','snippet','teaser_text','teaser_title']

for col in cols:
    zeit_df[col]=zeit_df[col].apply(lambda x: lemmatizer(x))
```

```python
zeit_df["newspaper"]=zeit_df["teaser_title"].apply(lambda x: "Die Zeit")
zeit_df["headline"]=zeit_df["title"]+" "+zeit_df["snippet"]+" "+zeit_df["teaser_text"]
```

```python
def hrefs_ids(href):
    first_part = href.split("http://www.zeit.de/")
    id_tag = first_part[1].split('/')
    return id_tag[0]

zeit_df["channel"]=zeit_df["url"].apply(lambda x: hrefs_ids(x))
```

```python
counts=zeit_df["channel"].value_counts(normalize=True)

print("Before deleting channels: ", zeit_df.shape)
zeit_df = zeit_df[~zeit_df['channel'].isin(counts[counts < 0.025].index)]
print("After deleting channels: ", zeit_df.shape)
```

```python
print("Before deleting channel 'Sport': ", zeit_df.shape)
zeit_df = zeit_df[~zeit_df['channel'].isin(["Sport"])]
print("After deleting channel 'Sport': ", zeit_df.shape)
```

### Applying corona score

"corona_dictionary_german.csv" will be updated every week with the latest newspaper articels scrapped.

```python
keywords=pd.read_csv("/Users/carolinvogt/Becoming_Data_Analyst/vaccine_sentiment/german/corona_dictionary_german.csv")
```

```python
corona_list=["corona","rki","wirkstoff","drosten","wieler","merkel","spahn","coronavirus","virus","lockdown","pandemie","epidemie","pcr-test","impfzentren","hausärzte","schnelltest","teststrategie","intensivbetten","covid","covid-19","impfung","impfstrategie","impfen","biontech","pfizer", "biontech/pfizer","moderna","johnson&johnson", "curevac","astrazeneca","oxford","janssen"]

def corona_related(x):
    temp_list=x.split(' ')

    score=0
    
    for word in temp_list:
        #print(word)
        if word in corona_list:
            score_corona=1
            #print(score)
            return score_corona
        
        elif word in list(keywords["keyword"]):
            score+=1
            #print(score)
        else:
            continue
            
    return score/(len(temp_list))

zeit_df["corona_score"]=zeit_df["headline"].apply(lambda x: corona_related(x))
```

```python
print("Before deleting non-corona-related articels: ", zeit_df.shape)
zeit_df=zeit_df[zeit_df["corona_score"]>0.5]
print("After deleting non-corona-related articels: ", zeit_df.shape)
```

### Applying vaccine dictionary

```python
vaccine=["mrna","impfquote","geimpfte","impfzentrum","impfzentren","impfung","impfen","impfstrategie","wirkstoff","biontech","pfizer", "biontech/pfizer","moderna","johnson&johnson", "curevac","astrazeneca","astra-zeneca","oxford","janssen"]


def vaccine_mentioned(x):
    temp_list=x.split(' ')
    
    for word in temp_list:
        #print(word)
        if word in vaccine:
            return word

        else:
            return "n/a"
        
zeit_df["vaccine"]=zeit_df["headline"].apply(lambda x: vaccine_mentioned(x))
```
